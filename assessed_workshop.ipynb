```python
import matplotlib.pyplot as plt
import matplotlib
import sqlite3
# matplotlib.use('Qt5Agg')
import pandas as pd
years, co2, temp = [], [], []
connection = sqlite3.connect("climate.db")
cursor = connection.cursor()

cursor.execute("SELECT Year FROM ClimateData")
years_info = cursor.fetchall()
for _info in years_info:
    years.append(_info[0])

cursor.execute("SELECT CO2 FROM ClimateData")
co2_info = cursor.fetchall()
for _info in co2_info:
    co2.append(_info[0])

cursor.execute("SELECT Temperature FROM ClimateData")
temp_info = cursor.fetchall()
for _info in temp_info:
    temp.append(_info[0])

plt.subplot(2, 1, 1)
plt.plot(years, co2, 'b--')
plt.title("Climate Data")
plt.ylabel("[CO2]")
plt.xlabel("Year (decade)")
plt.show()

# ===========================================================
climate = pd.read_csv("climate.csv")
years = climate['Year']
co2 = climate['CO2']
temp = climate['Temperature']

plt.subplot(2, 1, 1)
plt.plot(years, co2, 'b--')
plt.title("Climate Data")
plt.ylabel("[CO2]")
plt.xlabel("Year (decade)")

plt.subplot(2, 1, 2)
plt.plot(years, temp, 'r*-')
plt.ylabel("Temp (C)")
plt.xlabel("Year (decade)")
plt.show()
plt.savefig("co2_temp.png")

```

```python
import matplotlib.pyplot as plt
import sqlite3
database = sqlite3.connect("./climate.db")
cursor = database.cursor()
cursor.execute("SELECT * FROM ClimateData")
results = cursor.fetchall()
years = []
co2 = []
temp = []

for row in results:
    years.append(row[0])
    co2.append(row[1])
    temp.append(row[2])
plt.subplot(2, 1, 1)
plt.plot(years, co2, 'b--') 
plt.title("Climate Data") 
plt.ylabel("[CO2]") 
plt.xlabel("Year (decade)") 

plt.subplot(2, 1, 2)
plt.plot(years, temp, 'r*-') 
plt.ylabel("Temp (C)") 
plt.xlabel("Year (decade)") 
plt.show() 
plt.savefig("co2_temp_1.png") 
```
