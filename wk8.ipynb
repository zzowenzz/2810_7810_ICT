## Python for Excel

```python
import openpyxl
from openpyxl.chart import AreaChart,Reference

# create a workbook
wb = openpyxl.Workbook()
# get the active worksheet
ws = wb.active

# create a table
rows = [
['Number', 'Batch 1', 'Batch 2'],
[2, 40, 30],
[3, 40, 25],
[4, 50, 30],
[5, 30, 10],
[6, 25, 5],
[7, 50, 10],
]

# write the table to the worksheet
for row in rows:
    ws.append(row)

# create a chart
chart = AreaChart()
chart.title = "Area Chart"
chart.style = 13
chart.x_axis.title = 'Test'
chart.y_axis.title = 'Percentage'

# add the data to the chart
cats = Reference(ws, min_col=1, min_row=1,max_row=7) # create a reference to the first column (Number) from row 1 to 7
data = Reference(ws, min_col=2, min_row=1,max_col=3, max_row=7) # create a reference to the second and third column (Batch 1 and Batch 2) from row 1 to 7
chart.add_data(data, titles_from_data=True) # add the data to the chart
chart.set_categories(cats)
ws.add_chart(chart, "A10")
wb.save("./data/area.xlsx")
```

## Regular expression

```python
import re

txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt) # search for a string that starts with "The", ends with "Spain", and has any number of characters in between
print(x)

txt = "The rain in Spain, 2020"
re_pattern = "[a-e0-9]" # find all lower case characters a to e and numbers 0 to 9
x = re.findall(re_pattern, txt) # find all occurrences of the pattern in the string
print(x)
```

```python
txt = "The rain in Spain, 2020"
re_pattern ="\d\Z" # find all numbers at the end of the string; \d: any digit; \Z: end of string
x = re.search(re_pattern, txt) # find all occurrences of the pattern in the string
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

## Numpy

Numpy is a Python library for scientific computing. It provides a high-performance multidimensional array object, and tools for working with these arrays.

```python
import numpy as np

x = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print(x.ndim) # number of dimensions
print(x.shape) # shape of the array
print(np.linspace(1., 4., 6)) # create an array of 6 numbers from 1 to 4
print(np.random.random((2,3))) # create a 2 by 3 array of random numbers
```

```python
print(np.arange(3)) # create an array of 3 numbers from 0 to 2
print(np.arange(9).reshape(3,3)) # create a 3 by 3 array of numbers from 0 to 8
print(np.arange(8).reshape(2,2,2)) # create a 2 by 2 by 2 array of numbers from 0 to 7

x = np.arange(9)
print(x)
print(x[2])
print(x[-2])
```

## Matplotlib

Matplotlib is a Python library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt

# our X values:
days = list(range(0, 22, 3))
# our Y values:
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]

plt.plot(days, celsius_values)
plt.show()

plt.plot(days, celsius_values, 'bo')
plt.show()
```

```python
import matplotlib.pyplot as plt

X = np.linspace(0, 2 * np.pi, 50, endpoint=True)
F1 = 3 * np.sin(X)
F2 = np.sin(2*X)
F3 = 0.3 * np.sin(X)
F4 = np.cos(X)

plt.plot(X, F1, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, F2, color="red", linewidth=1.5, linestyle="--")
plt.plot(X, F3, color="green", linewidth=2, linestyle=":")
plt.plot(X, F4, color="grey", linewidth=2, linestyle="-.")
plt.legend(['F1','F2','F3','F4'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(1,11,100)

# the histogram of the data
plt.hist(x, bins=9, density=False, color='g')
plt.xlabel('score')
plt.ylabel('count')
plt.grid(True, axis='x')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)
plt.plot (X, Y, color='blue', alpha=1.00)
plt.fill_between(X, Y, y2=0, color='blue', alpha=.1)
plt.show()
```
